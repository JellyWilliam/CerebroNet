# Групповой анализ

```python
# Группировка по столбцу "death01"
df_group_death = df.groupby("death01")

# Разделение выборки на death01 == 0 и 1
df_death_0 = df[df["death01"] == 0]
df_death_1 = df[df["death01"] == 1]
```

```python
def group_analysis(column):
    print("Модальное значение в death01=0 :", df_death_0[column].mode().values[0], "\nМодальное значение в death01=1 :", df_death_1[column].mode().values[0])
    print("Распределение количества значений в столбце\n", df_group_death[column].value_counts())
    return df_group_death[column].describe()
```

## RNASeqCluster

```python
group_analysis("RNASeqCluster")
```

***Вывод:*** заметно преобладание RNASeqCluster = 4 для тех, кто не умер, и RNASeqCluster = 2 для тех, кто умер

## MethylationCluster

```python
group_analysis("MethylationCluster")
```
